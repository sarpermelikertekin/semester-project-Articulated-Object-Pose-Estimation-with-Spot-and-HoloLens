# Bounding Box and Keypoint Validator

This tool can be used to validate the annotations of the images

```python
from PIL import Image, ImageDraw, ImageFont

def draw_bounding_box_and_keypoints_from_txt(image_path, txt_path):
    x_norm = 875
    y_norm = 492
    
    # Open the provided image
    image = Image.open(image_path)
    draw = ImageDraw.Draw(image)
    
    # Load data from the txt file
    with open(txt_path, 'r') as file:
        # Assuming the structure is consistent with the example provided
        line = file.readline().strip()
        parts = line.split(' ')
        
        # Extracting bounding box information
        class_label = parts[0]  # Not used in drawing but might be useful for other purposes
        center_x, center_y = float(parts[1]), float(parts[2])
        width, height = float(parts[3]), float(parts[4])
        
        # Drawing the bounding box
        top_left_x = center_x * x_norm - (width * x_norm) / 2
        top_left_y = center_y * y_norm - (height * y_norm) / 2
        bottom_right_x = center_x * x_norm + (width * x_norm) / 2
        bottom_right_y = center_y * y_norm + (height * y_norm) / 2
        draw.rectangle([top_left_x, top_left_y, bottom_right_x, bottom_right_y], outline="black")
        
        # Draw the center point
        center_point_radius = 5
        draw.ellipse([center_x * x_norm - center_point_radius, center_y * y_norm - center_point_radius, 
                      center_x * x_norm + center_point_radius, center_y * y_norm + center_point_radius], 
                     fill="blue", outline="blue")
        
        # Drawing keypoints
        keypoints = parts[5:]
        for i in range(0, len(keypoints), 3):
            x, y, is_visible = float(keypoints[i]), float(keypoints[i+1]), int(keypoints[i+2])
            draw_keypoint(draw, x * x_norm, y * y_norm, is_visible, i // 3)
    
    # Save or show the image
    output_path = 'annotated_' + image_path.split('/')[-1]
    # image.save(output_path)
    image.show()  # Uncomment to display the image

def draw_keypoint(draw, x, y, is_visible, index):
    # Draw the index of the keypoint
    font_size = 20
    try:
        font = ImageFont.truetype("arial.ttf", font_size)
    except IOError:
        font = ImageFont.load_default()
    text_offset_x, text_offset_y = font.getsize(str(index))
    draw.text((x - text_offset_x / 2, y - text_offset_y / 2), str(index), fill="green" if is_visible == 2 else "red", font=font)

# Example usage
image_path = 'C:/Users/sakar/Semester Project/Spot Datasets/test/images/2.png'  # Replace with your image filename
txt_path = 'C:/Users/sakar/Semester Project/Spot Datasets/test/labels/2.txt'  # Replace with your TXT filename

draw_bounding_box_and_keypoints_from_txt(image_path, txt_path)
```

# Bounding Box and Keypoints Mask

This tool can be used to generate mask for Bounding Box and Keypoints

Press **ESC** key to stop the process

```python
import cv2
import numpy as np
import json
import os
import random

def load_json(json_path):
    with open(json_path, 'r') as file:
        data = json.load(file)
    return data

def calculate_bounding_box(image, center, size):
    height, width, _ = image.shape
    x_center, y_center = center['x'] * width, center['y'] * height
    box_width, box_height = size['x'] * width, size['y'] * height
    top_left = (int(x_center - box_width / 2), int(y_center - box_height / 2))
    bottom_right = (int(x_center + box_width / 2), int(y_center + box_height / 2))
    return top_left, bottom_right

def create_canvas_with_keypoints(image_dimensions, keypoints, color=(36, 236, 255)):
    canvas = np.zeros((image_dimensions[0], image_dimensions[1], 3), dtype=np.uint8)
    for kp in keypoints:
        position = (int(kp['position']['x'] * image_dimensions[1]), int(kp['position']['y'] * image_dimensions[0]))
        cv2.circle(canvas, position, radius=5, color=color, thickness=cv2.FILLED)
    return canvas

def create_canvas_with_bounding_box(image_dimensions, top_left, bottom_right, color=(36, 236, 255)):
    canvas = np.zeros((image_dimensions[0], image_dimensions[1], 3), dtype=np.uint8)
    cv2.rectangle(canvas, top_left, bottom_right, color, thickness=cv2.FILLED)
    return canvas

def process_image(image_path, json_path, base_path, dataset_name, save_images=False, image_id=None):
    image = cv2.imread(image_path)
    if image is None:
        raise FileNotFoundError(f"Image not found at specified path: {image_path}.")
    data = load_json(json_path)
    
    image_dimensions = image.shape
    keypoints_image = create_canvas_with_keypoints(image_dimensions, data['keypoints'])
    center = data['boundingBoxCenter']
    size = data['boundingBoxSize']
    top_left, bottom_right = calculate_bounding_box(image, center, size)
    bbox_image = create_canvas_with_bounding_box(image_dimensions, top_left, bottom_right)
    
    analysis_path = os.path.join(base_path, "Analysis", dataset_name)
    keypoints_path = os.path.join(analysis_path, "Keypoint Masks")
    bbox_path = os.path.join(analysis_path, "Bounding Box Masks")
    
    # Create the directories if they don't exist
    os.makedirs(keypoints_path, exist_ok=True)
    os.makedirs(bbox_path, exist_ok=True)

    if save_images and image_id is not None:
        # Use image_id to generate the filenames
        keypoints_image_path = os.path.join(keypoints_path, f'{image_id}_keypoints.png')
        bbox_image_path = os.path.join(bbox_path, f'{image_id}_bbox.png')
        
        cv2.imwrite(keypoints_image_path, keypoints_image)
        cv2.imwrite(bbox_image_path, bbox_image)
    
    # # Optionally display images
    # cv2.imshow("Keypoints Image", keypoints_image)
    # cv2.imshow("Filled Bounding Box Image", bbox_image)
    # while True:
    #     if cv2.waitKey(1) & 0xFF == 27:  # Correctly use bitwise AND to check if ESC is pressed
    #         break
    # cv2.destroyAllWindows()

    
def select_and_process_random_images(base_path, dataset_name, n=20):
    # Path to the images directory
    images_dir = os.path.join(base_path, dataset_name, "train", "images")
    # List all image files
    all_images = [img for img in os.listdir(images_dir) if img.endswith('.png')]
    
    # Ensure we don't select more images than available
    n = min(n, len(all_images))
    
    # Randomly select n images
    selected_images = random.sample(all_images, n)
    
    # Process each selected image
    for img_filename in selected_images:
        image_id = img_filename.split('.')[0]
        image_path = os.path.join(images_dir, img_filename)
        json_path = os.path.join(base_path, dataset_name, "train", "mapping_2d", f"{image_id}.json")
        
        # Call process_image for each selected image
        process_image(image_path, json_path, base_path, dataset_name, save_images=True, image_id=image_id)
        cv2.destroyAllWindows()  
        
# Example usage
base_path = "C:/Users/sakar/Semester Project/Spot Datasets"
dataset_name = "11 - Spot and AnyMAL"
select_and_process_random_images(base_path, dataset_name, n=20)
```

# Bounding Box and Keypoints Histogram

This tool can be used to generate histograms for Bounding Box and Keypoints

```python
import os
import matplotlib.pyplot as plt
import numpy as np

# Path to the directory containing the .txt files
base_path = r"C:\Users\sakar\Semester Project\Spot Datasets"
dataset_name = "11 - Spot and AnyMAL"

# Construct the directory path for reading
directory_path = os.path.join(base_path, dataset_name, "train", "labels")

# Construct the save directory path for writing
save_dir = os.path.join(base_path, "Analysis", dataset_name, "Data Analysis")

# Ensure the save directory exists
os.makedirs(save_dir, exist_ok=True)

# Initialize lists to store data from all files
all_bb_positions = []
all_bb_sizes = []
all_keypoints = []  # This will now dynamically adjust to the number of keypoints

# Function to parse a single line of data
def parse_data_line(line):
    data_elements = line.split()
    bb_positions = []
    bb_sizes = []
    keypoints = []  # Temporarily store keypoints for a single line

    i = 0
    while i < len(data_elements):
        # The class_id is used to identify the start of a new entry
        class_id = int(data_elements[i])
        i += 1

        # Extract and store bounding box center, width, and height
        bb_center_x, bb_center_y = float(data_elements[i]), float(data_elements[i+1])
        bb_width, bb_height = float(data_elements[i+2]), float(data_elements[i+3])
        bb_positions.append((bb_center_x, bb_center_y))
        bb_sizes.append((bb_width, bb_height))
        i += 4

        # Key points processing
        while i < len(data_elements) and not data_elements[i].isdigit():
            keypoint = [float(data_elements[i]), float(data_elements[i+1]), int(data_elements[i+2])]
            keypoints.append(keypoint)
            i += 3
    
    return bb_positions, bb_sizes, keypoints

# Iterate through each file in the directory
for filename in os.listdir(directory_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(directory_path, filename)
        
        with open(file_path, 'r') as file:
            for line in file:
                bb_positions, bb_sizes, keypoints = parse_data_line(line)
                
                # Append bounding box data
                all_bb_positions.extend(bb_positions)
                all_bb_sizes.extend(bb_sizes)

                # Adjust and append keypoint data
                for i, kp in enumerate(keypoints):
                    # Ensure there's a sublist for each keypoint
                    if i >= len(all_keypoints):
                        all_keypoints.append([])
                    all_keypoints[i].append(kp)

def plot_bb_positions_on_canvas(bb_positions, canvas_width=1600, canvas_height=900):
    # Adjust the height to include space for the yellow strip
    yellow_strip_height = 50  # Height of the yellow strip
    total_canvas_height = canvas_height + yellow_strip_height

    # Create a black canvas including space for the yellow strip
    canvas = np.zeros((total_canvas_height, canvas_width, 3), dtype=np.uint8)

    fig, ax = plt.subplots()
    ax.imshow(canvas, aspect='auto')
    
    # Scale the normalized positions to the canvas size
    scaled_positions = [(x * canvas_width, y * canvas_height) for x, y in bb_positions]

    # Plot white circles for each bounding box position
    for x, y in scaled_positions:
        circle = plt.Circle((x, y), 10, color='white', fill=True)  # Use white color for circles
        ax.add_artist(circle)

    # Add a yellow strip at the bottom
    yellow_strip = plt.Rectangle((0, canvas_height), canvas_width, yellow_strip_height, color='yellow')
    ax.add_artist(yellow_strip)

    # Add text on the yellow strip
    ax.text(canvas_width / 2, canvas_height + yellow_strip_height / 2, 'Distribution of Bounding Box Centers',
            color='black', ha='center', va='center')

    ax.set_xlim([0, canvas_width])
    ax.set_ylim([total_canvas_height, 0])  # Adjusted for the yellow strip
    plt.axis('off')  # Hide axes
    
    save_path = os.path.join(save_dir, 'distribution_of_bb_centers.png')
    plt.savefig(save_path)
    plt.show()
    plt.close()

def plot_bb_centers_heatmap(bb_positions, canvas_width=1600, canvas_height=900, bins=50, cmap='viridis'):
    # Scale the normalized positions to the canvas size
    scaled_positions = [(x * canvas_width, y * canvas_height) for x, y in bb_positions]

    # Extract scaled x and y values
    x_values, y_values = zip(*scaled_positions)

    plt.figure(figsize=(10, 6))
    plt.hist2d(x_values, y_values, bins=bins, cmap=cmap, cmin=1)
    plt.colorbar()
    plt.title('Heatmap of Bounding Box Centers')
    plt.xlabel('Position X')
    plt.ylabel('Position Y')
    plt.grid(False)  # Turn off the grid for clarity
    
    save_path = os.path.join(save_dir, 'heatmap_of_bb_centers.png')
    plt.savefig(save_path)
    plt.show()
    plt.close()
    
def plot_bb_areas_histogram(bb_sizes, resolution_width=1600, resolution_height=900, bins=50):
    # Calculate the area of each bounding box in pixels
    bb_areas = [(w * resolution_width) * (h * resolution_height) for w, h in bb_sizes]

    # Plot the histogram of bounding box areas
    plt.figure(figsize=(10, 6))
    plt.hist(bb_areas, bins=bins, color='blue', edgecolor='black', alpha=0.7)
    plt.title('Histogram of Bounding Box Areas')
    plt.xlabel('Area in pixels')
    plt.ylabel('Frequency')
    plt.grid(axis='y', alpha=0.75)
    
    save_path = os.path.join(save_dir, 'histogram_of_bb_areas.png')
    plt.savefig(save_path)
    plt.show()
    plt.close()

def plot_keypoints_distribution(all_keypoints, canvas_width=1600, canvas_height=900):
    yellow_strip_height = 50  # Height of the yellow strip
    total_canvas_height = canvas_height + yellow_strip_height

    for i, keypoints in enumerate(all_keypoints):
        # Create a black canvas including space for the yellow strip
        canvas = np.zeros((total_canvas_height, canvas_width, 3), dtype=np.uint8)

        fig, ax = plt.subplots()
        ax.imshow(canvas, aspect='auto')

        # Plot each keypoint according to its visibility flag
        for x, y, visibility_flag in keypoints:
            # Scale the normalized positions to the canvas size
            scaled_x, scaled_y = x * canvas_width, y * canvas_height
            
            if visibility_flag == 1:
                color = 'red'  # Visible keypoints
            else:
                color = 'green'  # Keypoints with a flag of 2
                
            circle = plt.Circle((scaled_x, scaled_y), 10, color=color, fill=True)
            ax.add_artist(circle)

        # Add a yellow strip at the bottom
        yellow_strip = plt.Rectangle((0, canvas_height), canvas_width, yellow_strip_height, color='yellow')
        ax.add_artist(yellow_strip)

        # Add text on the yellow strip indicating the keypoint index
        ax.text(canvas_width / 2, canvas_height + yellow_strip_height / 2, f'Distribution of Keypoint {i}',
                color='black', ha='center', va='center')

        ax.set_xlim([0, canvas_width])
        ax.set_ylim([total_canvas_height, 0])  # Adjusted for the yellow strip
        plt.axis('off')  # Hide axes
        
        save_path = os.path.join(save_dir, f'distribution_of_keypoint_{i}.png')
        plt.savefig(save_path)
        plt.show()
        plt.close()
        
def plot_keypoints_xy_histograms(all_keypoints, bins=50):
    for i, keypoints in enumerate(all_keypoints):
        # Extract x and y values from keypoints, ignoring the visibility flag
        x_values = [kp[0] for kp in keypoints]
        y_values = [kp[1] for kp in keypoints]

        # Plot histogram for x values
        plt.figure(figsize=(10, 6))
        plt.hist(x_values, bins=bins, color='skyblue', edgecolor='black', alpha=0.7)
        plt.title(f'Keypoint {i} - X Values Distribution')
        plt.xlabel('Normalized X values')
        plt.ylabel('Frequency')
        plt.grid(axis='y', alpha=0.75)        
        save_path = os.path.join(save_dir, f'histogram_of_keypoint_{i}_x_distribution.png')
        plt.savefig(save_path)
        plt.show()
        plt.close()

        # Plot histogram for y values
        plt.figure(figsize=(10, 6))
        plt.hist(y_values, bins=bins, color='salmon', edgecolor='black', alpha=0.7)
        plt.title(f'Keypoint {i} - Y Values Distribution')
        plt.xlabel('Normalized Y values')
        plt.ylabel('Frequency')
        plt.grid(axis='y', alpha=0.75)
        
        save_path = os.path.join(save_dir, f'histogram_of_keypoint_{i}_y_distribution.png')
        plt.savefig(save_path)
        plt.show()
        plt.close()
        
def plot_keypoints_heatmap(all_keypoints, bins=50, cmap='hot'):
    for i, keypoints in enumerate(all_keypoints):
        # Extract x and y values from keypoints, ignoring the visibility flag
        x_values = [kp[0] for kp in keypoints]
        y_values = [kp[1] for kp in keypoints]

        plt.figure(figsize=(8, 6))
        plt.hist2d(x_values, y_values, bins=bins, cmap=cmap, cmin=1)
        plt.colorbar()
        plt.title(f'Heatmap of Keypoint {i}')
        plt.xlabel('Normalized X values')
        plt.ylabel('Normalized Y values')
        plt.grid(False)
        
        save_path = os.path.join(save_dir, f'heatmap_of_keypoint_{i}.png')
        plt.savefig(save_path)
        plt.show()
        plt.close()

plot_bb_positions_on_canvas(all_bb_positions)
plot_bb_centers_heatmap(all_bb_positions)
plot_bb_areas_histogram(all_bb_sizes)
plot_keypoints_distribution(all_keypoints)
plot_keypoints_xy_histograms(all_keypoints)
plot_keypoints_heatmap(all_keypoints)
```
